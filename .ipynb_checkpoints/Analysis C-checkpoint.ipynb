```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
from scipy import stats
import numpy as np


def save_obj(obj, name, path = 'obj/'):
    with open(path + name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name, path = 'obj/'):
    with open(path + name + '.pkl', 'rb') as f:
        return pickle.load(f)

datadir = '00Taken'
participants = load_obj('participants')
```

## Part 1: NASA Raw TLX
Generate NASA Raw TLX plots as used in paper.

```python
# tlx_df = pd.read_csv(f'{datadir}/Questionnaires/NASATLX_df.csv', index_col = 0)
# tlx_df

#load nasa tlx dataframe
tlx_df = pd.read_csv('obj/NASATLX_df.csv', index_col = 0)

keys = ["wl_noucfog", "wl_uctun", "wl_nouctun","wl_ucrain",  "wl_noucrain"]
out = pd.DataFrame({'Raw TLX':[], 'condition':[]})

for key in keys:
    vals = tlx_df[key]
    df = pd.DataFrame({'Raw TLX':vals, 'condition':[key]*len(vals)})
    out = pd.concat([out,df])
    
out = out.replace(["wl_noucfog", "wl_uctun", "wl_nouctun","wl_ucrain",  "wl_noucrain"],
            ['MC-HU', 'MU-HU-uc', 'MU-HU', 'MU-HC-uc', 'MU-HC']) #replace labels

plt.subplots(figsize = (5,3))
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")

ax = sns.boxplot(x="condition", y='Raw TLX', data=out, showfliers = False, color = 'skyblue')
ax = sns.swarmplot(x="condition", y="Raw TLX", data=out, color=".25")
ax.set_xlabel('')
ax.set_ylim(0,100)
plt.tight_layout()

```

### Statistical evaluation 

### of difference between the three groups (using Friedman):

```python
friedmanUFL = stats.friedmanchisquare(vdl_df['noucfog_ufl'], vdl_df['uctun_ufl'], vdl_df['ucrain_ufl'])
friedmanSAT = stats.friedmanchisquare(vdl_df['noucfog_sat'], vdl_df['uctun_sat'], vdl_df['ucrain_sat'])
```

```python
s=   f'''A Friedman test for the usefulness and satisfaction
indicated a significant difference between the conditions,
(Ï‡2(2) = {friedmanUFL[0]:.2f}, p < .001 for usefulness 
and Ï‡2(2) = {friedmanSAT[0]:.2f}, p = {friedmanSAT[1]:.4f} for satisfaction). 

    '''
print(s)
```

### Post-hoc Wilcoxon signed rank test:


```python
wil_ufl_foggyroad_rain = stats.wilcoxon(vdl_df['noucfog_ufl'], vdl_df['ucrain_ufl'])
wil_ufl_tun_foggyroad = stats.wilcoxon(vdl_df['uctun_ufl'], vdl_df['noucfog_ufl'])
wil_ufl_tun_rain = stats.wilcoxon(vdl_df['uctun_ufl'], vdl_df['ucrain_ufl'])

wil_sat_foggyroad_rain = stats.wilcoxon(vdl_df['noucfog_sat'], vdl_df['ucrain_sat'])
wil_sat_tun_foggyroad = stats.wilcoxon(vdl_df['uctun_sat'], vdl_df['noucfog_sat'])
wil_sat_tun_rain = stats.wilcoxon(vdl_df['uctun_sat'], vdl_df['ucrain_sat'])
```

```python
# Note that p-values are halved, as we test one-sided
print ('Usefulnesss:')
print('foggyroad vs rain: T = ', wil_ufl_foggyroad_rain[0], ', p = ', wil_ufl_foggyroad_rain[1]/2 )
print('tunnel vs foggyroad:',wil_ufl_tun_foggyroad[0], ', p = ', wil_ufl_tun_foggyroad[1]/2 )
print('tunnel vs rain:', wil_ufl_tun_rain[0], ', p = ', wil_ufl_tun_rain[1]/2 )

```

```python
# Note that p-values are halved, as we test one-sided
print ('Satisfaction:')
print('foggyroad vs rain: T = ', wil_sat_foggyroad_rain[0], ', p = ', wil_sat_foggyroad_rain[1]/2 )
print('tunnel vs foggyroad:',wil_sat_tun_foggyroad[0], ', p = ', wil_sat_tun_foggyroad[1]/2 )
print('tunnel vs rain:', wil_sat_tun_rain[0], ', p = ', wil_sat_tun_rain[1]/2 )
```
