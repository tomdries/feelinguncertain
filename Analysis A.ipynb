# Analysis A: MTTC analysis and raw data plots

This document contains the plotting and statistical analysis of the MTTC scores, and the generation of the raw data plot that displayed TTC, velocity, steer angle, brake depression and lane changing information.


```python
# Define functions and load data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from TomsFunctions import *

# def save_obj(obj, name, path = 'obj/'):
#     with open(path + name + '.pkl', 'wb') as f:
#         pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

# def load_obj(name, path = 'obj/'):
#     with open(path + name + '.pkl', 'rb') as f:
#         return pickle.load(f)
    
def MTTCscore(MTTCdf, method = 'median', inverse = False):
    # This function calculates the MTTC score based on the three trials.
    # Function returns the MTTC score based on the mean or median. 
    # When inverse = True, the inverse MTTC is returned.
    # method is 'mean' or 'median'
    
    MTTCscore = {'participant':[], 'score': []}
    for p in MTTCdf['participant'].unique(): #for participants
        currentlist = MTTCdf[MTTCdf['participant'] == p].iloc[:,1]
        MTTCscore['participant'].append(p)
        if method == 'median':
            MTTCscore['score'].append(currentlist.median())
        if method == 'mean':
            MTTCscore['score'].append(currentlist.mean())
    
    dfout = pd.DataFrame(MTTCscore)
    if inverse:
        dfout = pd.concat([dfout['participant'], dfout['score'].apply(lambda x: 1/x)], axis = 1)
    return dfout

def plot_scores(nouctun_scores, uctun_scores, noucfog_scores, title):
    # add column with run
    nouctun_scores = pd.concat([nouctun_scores, pd.DataFrame({'run': ['nouctun']*len(nouctun_scores)})], axis = 1)
    uctun_scores = pd.concat([uctun_scores, pd.DataFrame({'run': ['uctun']*len(uctun_scores)})], axis = 1)
    noucfog_scores = pd.concat([noucfog_scores, pd.DataFrame({'run': ['foggyroad']*len(noucfog_scores)})], axis = 1)

    merged = pd.concat([nouctun_scores, uctun_scores, noucfog_scores])
    print(merged)
    # Usual boxplot
    
    fig, ax = plt.subplots(figsize = (6,2))
    sns.set_context('paper', font_scale=1.5)

#     plt.tight_layout()
    sns.set_style('whitegrid')
    sns.boxplot(y='run', x='score', data=merged, ax=ax, color = 'skyblue')

    # Add jitter with the swarmplot function.
    sns.swarmplot(y='run', x='score', data=merged, color=".25", ax = ax)
    plt.yticks([0,1,2],['MU-HU','MU-HU-uc','MC-HU'])
    plt.xlabel('MTTC (s)')
    plt.tight_layout()
    return merged

def plot_differences(merged_df, runA, runB, binwidth):
    scoreA = merged_df.loc[merged_df['run'] == runA, 'score']
    scoreB = merged_df.loc[merged_df['run'] == runB, 'score']
    
    diff = scoreA - scoreB
    fig, ax = plt.subplots()
    plt.hist(x = diff, bins=np.arange(min(diff), max(diff) + binwidth, binwidth))
    return diff
    

# Load MTTC scores per trial
MTTCnouctun = load_obj('MTTCnouctun') 
MTTCnoucfog = load_obj('MTTCnoucfog') 
MTTCuctun = load_obj('MTTCuctun')

# Load list of participants
participants = load_obj('participants')
```


## Caculation of MTTC scores and plotting them. 


This image generated here was used in the Results section of the paper.

```python
%matplotlib inline

method = 'mean'
nouctun_scores = MTTCscore(MTTCnouctun, method, inverse = False)
noucfog_scores = MTTCscore(MTTCnoucfog, method, inverse = False)
uctun_scores = MTTCscore(MTTCuctun, method, inverse = False)
noinv_df = plot_scores(nouctun_scores, uctun_scores, noucfog_scores, '')
plt.ylabel('')
plt.tight_layout()
# plt.savefig('figs/MTTCbox.eps')
```

# Paired t-test

We decided to use the paired t-test, as the difference between scores seem reasonably near the normal distribution, so we assume the sample distribution is as well. For completeness, we calculate cohen's d, where the denominator is the pooled standard deviation.

```python
from scipy import stats
from numpy import std, mean, sqrt

def cohen_d(x,y):
# denominator is pooled standard deviation which is generally only appropriate 
# if the population standard deviation is equal for both groups:
    nx = len(x)
    ny = len(y)
    dof = nx + ny - 2
    return (mean(x) - mean(y)) / sqrt(((nx-1)*std(x, ddof=1) ** 2 + (ny-1)*std(y, ddof=1) ** 2) / dof)

a = noinv_df.loc[noinv_df['run'] == 'nouctun', 'score']
b = noinv_df.loc[noinv_df['run'] == 'uctun', 'score']
c = noinv_df.loc[noinv_df['run'] == 'foggyroad', 'score']


Ttest_nouctun_uctun = stats.ttest_rel(a, b)
Ttest_uctun_foggyroad = stats.ttest_rel(b, c)
d_nouctun_uctun = cohen_d(a,b)
d_uctun_foggyroad = cohen_d(b,c)

print('MC-HU vs MU-HU-uc:', Ttest_nouctun_uctun, 'd =', d_nouctun_uctun)
print('MU-HU vs MU-HU-uc', Ttest_uctun_foggyroad, 'd =', d_uctun_foggyroad)
```
