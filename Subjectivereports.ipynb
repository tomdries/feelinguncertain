```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from scipy import stats


def save_obj(obj, name, path = 'obj/'):
    with open(path + name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name, path = 'obj/'):
    with open(path + name + '.pkl', 'rb') as f:
        return pickle.load(f)

datadir = '00Taken'
participants = load_obj('participants')
```

```python
friedmanUFL = stats.friedmanchisquare(vdl_df['noucfog_ufl'], vdl_df['uctun_ufl'], vdl_df['ucrain_ufl'])
friedmanSAT = stats.friedmanchisquare(vdl_df['noucfog_sat'], vdl_df['uctun_sat'], vdl_df['ucrain_sat'])
```

```python
s=   f'''A Friedman test for the usefulness and satisfaction
indicated a significant difference between the conditions,
(χ2(2) = {friedmanUFL[0]:.2f}, p < .001 for usefulness 
and χ2(2) = {friedmanSAT[0]:.2f}, p = {friedmanSAT[1]:.4f} for satisfaction). 

    '''
print(s)
```

```python
wil_ufl_foggyroad_rain = stats.wilcoxon(vdl_df['noucfog_ufl'], vdl_df['ucrain_ufl'])
wil_ufl_tun_foggyroad = stats.wilcoxon(vdl_df['uctun_ufl'], vdl_df['noucfog_ufl'])
wil_ufl_tun_rain = stats.wilcoxon(vdl_df['uctun_ufl'], vdl_df['ucrain_ufl'])

wil_sat_foggyroad_rain = stats.wilcoxon(vdl_df['noucfog_sat'], vdl_df['ucrain_sat'])
wil_sat_tun_foggyroad = stats.wilcoxon(vdl_df['uctun_sat'], vdl_df['noucfog_sat'])
wil_sat_tun_rain = stats.wilcoxon(vdl_df['uctun_sat'], vdl_df['ucrain_sat'])
```

```python
print ('Usefulnesss:')
print('foggyroad vs rain:', wil_ufl_foggyroad_rain)
print('tunnel vs foggyroad:', wil_ufl_tun_foggyroad)
print('tunnel vs rain:', wil_ufl_tun_rain)

```
